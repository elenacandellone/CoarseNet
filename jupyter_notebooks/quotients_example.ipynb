This method relies on the automorphism orbits of vertices, which are the sets of nodes such that there exists automorphisms amongst one another. All nodes in a given automorphism orbit play an identical structural role in the network, and are only distinguishable by having distinct node labels.

### Toy example (from the paper)

````python
g = nx.Graph()
g.add_edges_from([
    (0,1),
    (2,1),
    (3,1),
    (1,4),
    (4,5),
    (4,6),
    (4,7),
    (5,8),
    (7,9),
    (7,6),
    (5,6),
    (6,10),
    (6,11),
    (10,12),
    (10,13),
    (10,14),
    (11,12),
    (11,13),
    (11,14),
    (12,13),
    (12,14),
    (13,14)
])
````

````python
coarse_grained_net = get_coarse_grained_net(g, dev=True)

pos_g = {
    0: (0,0),
    1: (1,0),
    2: (0,1),
    3: (0,-1),
    4: (2,0),
    5: (2.5,-1),
    6: (3,0),
    7: (2.5,1),
    8: (2.5,-2),
    9: (2.5,2),
    10: (4,0.5),
    11: (4,-0.5),
    12: (5,0),
    13: (5,1),
    14: (5,-1)
}

pos_course = {
    15: (0,0),
    1: (1,0),
    4: (2,0),
    17: (2.5,-1),
    6: (3,0),
    18: (2.5,-2),
    19: (4,0),
    16: (5,0)
}

fig, axs = plt.subplots(1,2, figsize = (14,5))

node_colors = [v[1]['partition'] for v in g.nodes(data=True)]

nx.draw(g,pos = pos_g, node_color = node_colors, cmap = plt.cm.tab20, with_labels = True, ax=axs[0])
axs[0].set_title('Partitioned graph')

node_colors = [v[1]['partition'] for v in coarse_grained_net.nodes(data=True)]

nx.draw(coarse_grained_net, pos = pos_course, node_color=node_colors, cmap = plt.cm.tab20, with_labels = True, ax=axs[1])
axs[1].set_title('Coarsed grained graph')

axs[1].set_ylim(axs[0].get_ylim())

plt.show()
````
